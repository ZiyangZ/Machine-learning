#Load data

Use 80% data as training data and 20% as testing data.

```
from keras.datasets import cifar10

(train_images_original, train_labels_original), (test_images_original, test_labels_original) = cifar10.load_data()

train_images_all = train_images_original.astype('float32') / 255
test_images = test_images_original.astype('float32') / 255

from keras.utils import to_categorical

train_labels_all = to_categorical(train_labels_original)
test_labels = to_categorical(test_labels_original)

train_images = train_images_all[:40000,:]
validation_images = train_images_all[40000:,:]
train_labels = train_labels_all[:40000,:]
validation_labels = train_labels_all[40000:,:]
```

```
print(train_images.shape, train_labels.shape, validation_images.shape, validation_labels.shape, test_images.shape, test_labels.shape)
```

#Architecture 1

Basic architecture.

```
from keras import layers
from keras import models
from keras import optimizers

batch_size = 32
epochs = 100

model = models.Sequential()

model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D(2, 2))

model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model.compile(
    loss='categorical_crossentropy', 
    optimizer=optimizers.RMSprop(lr=1e-4),
    metrics=['acc'])

model.fit(train_images, train_labels,
              batch_size=batch_size,
              epochs=epochs,
              validation_data=(validation_images, validation_labels))
```

After 100 epochs, the validation accuracy is 0.6856

```
del model
```

#Architecture 2

Modify architecture 1 by adding additional layers.

```
from keras import layers
from keras import models
from keras import optimizers

batch_size = 32
epochs = 50

model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model.compile(
    loss='categorical_crossentropy', 
    optimizer=optimizers.RMSprop(lr=1e-4),
    metrics=['acc'])

model.fit(train_images, train_labels,
              batch_size=batch_size,
              epochs=epochs,
              validation_data=(validation_images, validation_labels))
              
```

The validation accuracy is slightly increased to 0.7253

```
del model
```

#Architecture 3

Add dropout layers to architecture 2.

```
from keras import layers
from keras import models
from keras import optimizers

batch_size = 32
epochs = 50

model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Dropout(0.25))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(10, activation='softmax'))

model.compile(
    loss='categorical_crossentropy', 
    optimizer=optimizers.RMSprop(lr=1e-4),
    metrics=['acc'])

model.fit(train_images, train_labels,
              batch_size=batch_size,
              epochs=epochs,
              validation_data=(validation_images, validation_labels))
```

The validation accuracy is increased to 0.7854.

```
del model
```

#Architecture 4

Add image data augmentation to architecture 3.

```
from keras.preprocessing.image import ImageDataGenerator
from keras import layers
from keras import models
from keras import optimizers

batch_size = 32
epochs = 100

model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Dropout(0.25))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(10, activation='softmax'))

model.compile(
    loss='categorical_crossentropy', 
    optimizer=optimizers.RMSprop(lr=1e-4),
    metrics=['acc'])

datagen = ImageDataGenerator(
    rotation_range=1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest')

datagen.fit(train_images)

model.fit_generator(datagen.flow(train_images, train_labels,
              batch_size=batch_size),
              steps_per_epoch = 1250,
              epochs=epochs,
              validation_data=(validation_images, validation_labels))

```

The validation accuracy is slightly decreased to 0.7658.

```
model.save('cifar10_model4.h5')
```

#Using architecture 4 to predict the test data

```
from keras.models import load_model

model = load_model('cifar10_model4.h5')
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(test_loss, test_acc)
```

The prediction accuracy on test data is 0.7608, which is very close to the validation accuracy 0.7658 in architecture 4.

```
del model
```

#Using 5-fold cross-validation to fit the training data using Architecture 4

```
from keras.preprocessing.image import ImageDataGenerator
from keras import layers
from keras import models
from keras import optimizers
from sklearn.model_selection import StratifiedKFold
import numpy as np

batch_size = 32
epochs = 100

seed = 8
np.random.seed(seed)

validate_accuracy = []
test_accuracy = []

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

for train, validate in kfold.split(train_images_all, train_labels_original):
  model = models.Sequential()

  model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
  model.add(layers.Conv2D(32, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D(2, 2))
  model.add(layers.Dropout(0.25))

  model.add(layers.Conv2D(64, (3, 3), activation='relu'))
  model.add(layers.Conv2D(64, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D(2, 2))
  model.add(layers.Dropout(0.25))

  model.add(layers.Flatten())
  model.add(layers.Dense(512, activation='relu'))
  model.add(layers.Dropout(0.25))
  model.add(layers.Dense(10, activation='softmax'))

  model.compile(
    loss='categorical_crossentropy', 
    optimizer=optimizers.RMSprop(lr=1e-4),
    metrics=['acc'])

  datagen = ImageDataGenerator(
    rotation_range=1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest')

  datagen.fit(train_images_all[train])

  model.fit_generator(datagen.flow(train_images_all[train], train_labels_all[train],
              batch_size=batch_size),
              steps_per_epoch = 1250,
              epochs=epochs,
              validation_data=(train_images_all[validate], train_labels_all[validate]))
  
  v_loss, v_acc = model.evaluate(train_images_all[validate], train_labels_all[validate])
  t_loss, t_acc = model.evaluate(test_images, test_labels)
  validate_accuracy.append(v_acc)
  test_accuracy.append(t_acc)

print(validate_accuracy, test_accuracy)
print(np.mean(validate_accuracy), np.mean(test_accuracy))
```

The prediction accuracy on validation data and test data are 0.7749 and 0.7721, respectively. The difference is smaller than the simple hold-out validation, which is 0.7658 and 0.7608, respectively.
